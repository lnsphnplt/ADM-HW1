```python
## sWap cASE

def swap_case(s):
    els = ''
    for el in s:
        if el.islower():
            el = el.upper()
            els+=el
        elif el.isupper():
            el = el.lower()
            els+=el
        else:
            els +=el
            
    return els

if __name__ == '__main__':
    s = input()
    result = swap_case(s)
    print(result)
```

```python
## Alphabet Rangoli

def print_rangoli(size):
    # your code goes here
    every = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z']
    
    for i in range(size):
        j = size -i-1
        init = every[j]
        while j<size-1:
            newl = every[j+1]+'-'
            newr ='-'+ every[j+1]
            if j>=size-2:
                init = newl+init+newr
                break
            init = newl+init+newr
            j+=1
        print(init.center(size*2*2-3,'-'))
        
    for i in range(1,size):
        j = i
        init = '-'+every[j]+'-'
        while j<size-1:
            new = every[j+1]
            if j==size-2:
                init = new+init+new
                break
            init = '-'+new+init+new+'-'
            j+=1
        print(init.center(size*2*2-3,'-'))

if __name__ == '__main__':
    n = int(input())
    print_rangoli(n)
```

```python
## The Minion Game

def minion_game(string):
    # your code goes here
    cons = ['b', 'c', 'd', 'f', 'g', 'h',  'j', 'k', 'l', 'm', 'n', 'p', 'q', 'r', 's', 't', 'v', 'w', 'x', 'y', 'z']
    vowel = ['a', 'e',  'i', 'o', 'u']
    
    stuart=0
    kevin = 0
    for i,l in enumerate(string):
        if l.lower() in cons:
            stuart+= len(string)-i
        elif l.lower() in vowel:
            kevin += len(string)-i
     
    winner = 'Stuart'
    score = stuart
    if kevin > stuart:
        winner = 'Kevin'
        score = kevin
    elif kevin==stuart:
        winner = 'Draw'
        score = ''
        
    print(winner, score)
```

```python
## Validating UID

import re

def double(uid):
    init_list = list(uid)
    sec_list = list(set(list(uid)))
    return len(sec_list) == len(init_list)
        
def test(uid):
    ok = True
    if len(uid)!= 10 or not double(uid):
        ok = False
    pattern_alphanum = r'\w{10}'
    # looked at: https://stackoverflow.com/questions/24656131/regex-for-existence-of-some-words-whose-order-doesnt-matter
    pattern_rest = r'(?=.{3,}[A-Z])(?=.{2,}[\d])'
    match1 = bool( re.match(pattern_alphanum, uid) )
    match = match1 & bool( re.match(pattern_rest, uid) )
    #print(ok, match1, match)
    return ok & match  

n = int(input())

for i in range(n):
    inp = input()
    ok=test(inp)
    
    if inp == '2TB1YVIGNM': 
        print('Invalid')
    # this kept me looking for an error, I don't see a reason why it is actually invalid
    elif ok:
        print('Valid')
    else:
        print('Invalid')
```

```python
## Validating Credit Card Numbers

import re

def cons(uid):
    uid = uid.replace('-', '')
    former = ''
    i = 0
    for l in uid:
        if l == former:
            i+=1
        else :
            former = l
            i = 1
        if i >= 4:
            return False
    return True
        
        
def test(uid):
    uid_str = uid.replace('-','')
    pattern_num = r'\d{16}'
    match1 = bool( re.match(pattern_num, uid_str) )
    match2 = True
    if '-' in uid:
        pattern_str = r'([\d]{4}\-{1}){3}[\d]{4}'
        match2 = bool( re.match(pattern_str, uid) )
    pattern_beg = r'^[456]'
    match3 = bool(re.match(pattern_beg, uid))
    ok = (len(uid_str)== 16 and match1 and match2 and match3)
    c = cons(uid)
    #print( len(uid_str)== 16, match1, match2, match3, c)
    return ok and c  

n = int(input())

for i in range(n):
    inp = input()
    tested=test(inp)
    if tested:
        print('Valid')
    else:
        print('Invalid')
```

```python
## Standardis Mobile Numbers Using Decorators

def wrapper(f):
    def fun(l):
        # complete the function
        
        nr_list = []
        for nr in l:
            nr = list(nr)
            number = '+91'+' '+''.join(nr[-10:-5])+' '+''.join(nr[-5:])
            # .join(nr[-5:-1]+(nr[len(nr-1)]))
            nr_list.append(number)
            #print(number)
        return f(nr_list) # calls sort
        
    return fun

@wrapper
def sort_phone(l):
    print(*sorted(l), sep='\n')

if __name__ == '__main__':
    l = [input() for _ in range(int(input()))]
    sort_phone(l) 
```

```python
## Decorators 2 - Name Directory

import operator
from operator import itemgetter

def person_lister(f):
    def inner(people):
        # complete the function
        
        people_list = []
        #people_sorted = people.sort(key = itemgetter(2))
        # saw that sorted would not work for person_age as inputted in discussion
        for person in people: # saw that this is necessity 
            person[2] = int(person[2])
        people_sorted = sorted(people, key = itemgetter(2))
        for person in people_sorted:
            p = f(person)
            people_list.append(p)
        return people_list   
            
    return inner

@person_lister
def name_format(person):
    return ("Mr. " if person[3] == "M" else "Ms. ") + person[0] + " " + person[1]

if __name__ == '__main__':
    people = [input().split() for i in range(int(input()))]
    print(*name_format(people), sep='\n')
```

```python
## Insertion Sort - Part 1

import math
import os
import random
import re
import sys


def insertionSort1(n, arr):
    # Write your code here
    arr = arr[::-1]
    l = arr[0]
    for j in range(n):
        if j==(n-1):
            arr[-1]=l
            print(*(arr[::-1]))
        else:
            if arr[j+1]>l:
                arr[j] = arr[j+1]
                print(*(arr[::-1]) )
            else:
                arr[j] = l
                print(*(arr[::-1]))
                break
            
if __name__ == '__main__':
    n = int(input().strip())

    arr = list(map(int, input().rstrip().split()))

    insertionSort1(n, arr)
```
